```python
#import your packages here

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Load the sales data here
data = pd.read_csv("sales_data.csv")
print(data.shape)
print(type(data))
#checking for null data
data.isnull().any()
print(data.head())
```

# Question 1

## i. The average sales and profit per market

```python
#Accessing the sales 
sales_data = data["Sales"]
sales_data.describe()
```

```python
#Accessing the market
market_data = data["Market"]
print(market_data.value_counts())
Number_of_Markets = len(market_data.value_counts())#Accessing the Number of Markets in the Data
Number_of_Markets
```

```python
#Accessing the Profit 
profit_data = data["Profit"]
print(profit_data.head())
profit_data.describe()
```

```python
## Average Sales and profit per salary
```

```python
#Average_sales_per_Salary = sum of sales/number of markets
Average_sales_per_Market =  sales_data.sum()/Number_of_Markets
print(f"The Average Sales per Market is : {round(Average_sales_per_Market,3)}") 
```

```python
#Average_Profit_per_Market = sum of profit/number of markets
Average_Profit_per_Market =  profit_data.sum()/Number_of_Markets
print(f"The Average Profit per Market is : {round(Average_Profit_per_Market ,3)}") 
```

### The Average Profit per Market is : 58133.604

## ii  stacked bar chart showing the sales and profit per market

```python
#Getting the Sum sales for the unique markets by accesing the indexes of each unique market and summing their sales
APAC_Sales = data.loc[data["Market"]=="APAC"].Sales.sum()
LATAM_Sales = data.loc[data["Market"]=="LATAM"].Sales.sum()
EU_Sales = data.loc[data["Market"]=="EU"].Sales.sum()
EMEA_Sales = data.loc[data["Market"]=="EMEA"].Sales.sum()
Africa_Sales = data.loc[data["Market"]=="Africa"].Sales.sum()
Canada_Sales = data.loc[data["Market"]=="Canada"].Sales.sum()
US_Sales = data.loc[data["Market"]=="US"].Sales.sum()


Sales_market = [APAC_Sales,LATAM_Sales,EU_Sales,EMEA_Sales,Africa_Sales,Canada_Sales,US_Sales]

#Getting the total profit for the unique markets by accesing the indexes of each unique market and summing their profit
APAC_Profit= data.loc[data["Market"]=="APAC"].Profit.sum()
LATAM_Profit = data.loc[data["Market"]=="LATAM"].Profit.sum()
EU_Profit = data.loc[data["Market"]=="EU"].Profit.sum()
EMEA_Profit = data.loc[data["Market"]=="EMEA"].Profit.sum()
Africa_Profit = data.loc[data["Market"]=="Africa"].Profit.sum()
Canada_Profit = data.loc[data["Market"]=="Canada"].Profit.sum()
US_Profit = data.loc[data["Market"]=="US"].Profit.sum()


Profit_market = [APAC_Profit,LATAM_Profit,EU_Profit,EMEA_Profit,Africa_Profit,Canada_Profit,US_Profit]
label = ["APAC",'LATAM',"EU","EMEA","Africa","Canada","US"]
```

```python

#plotting the stacked bar chart 
plt.figure(figsize=[12,8])
plt.title("stacked bar chart showing the sales and profit per market ", fontsize = 16)
plt.xlabel("Markets",fontsize=16)
#plot the Sales Per Market
plt.bar(label,Sales_market,label='Sales')

# Then plot the 'Profit' bars on top, starting at the top of the "Sales"
plt.bar(label,Profit_market,bottom = Sales_market, label='Profit')
plt.legend()
```

### iii Which market brings more sales on the average?

```python
#Getting the mean sales for the unique markets by accesing the indexes of each unique market and averaging their sales
APAC_Sales_mean = data.loc[data["Market"]=="APAC"].Sales.mean()
LATAM_Sales_mean = data.loc[data["Market"]=="LATAM"].Sales.mean()
EU_Sales_mean = data.loc[data["Market"]=="EU"].Sales.mean()
EMEA_Sales_mean = data.loc[data["Market"]=="EMEA"].Sales.mean()
Africa_Sales_mean = data.loc[data["Market"]=="Africa"].Sales.mean()
Canada_Sales_mean = data.loc[data["Market"]=="Canada"].Sales.mean()
US_Sales_mean = data.loc[data["Market"]=="US"].Sales.mean()


Average_Sales_market = [APAC_Sales_mean,LATAM_Sales_mean,EU_Sales_mean,
                              EMEA_Sales_mean,Africa_Sales_mean,Canada_Sales_mean,
                             US_Sales_mean]


```

### show the Average Sales Per Market on bar chart  to visualize the Average sales for each market

```python
#plot the Average Sales Per Market to visualize the Average sales for each market
plt.figure(figsize=[12,8])
plt.title("Bar chart showing the Average sales per market ", fontsize = 16)
plt.xlabel("Markets",fontsize=16)
plt.ylabel("Average Sales",fontsize=16)
plt.bar(label, Average_Sales_market,label='Sales')
 
```

### APAC Market bring brings more sales on the average

## i Extract the days and months from the Order Date and add them to the data frame with the name "Days" and "Months" respectively?

```python

data['Days']=(pd.to_datetime(data["Order Date"])).dt.day#getting the day
data["Months"]=(pd.to_datetime(data["Order Date"])).dt.month#getting the mont
data["Months_abrv"]=(pd.to_datetime(data["Order Date"])).dt.month_name().str[:3]#Getting the month in string(abrv) and adding it as a data column
```

```python
data
```

## ii Show the trend of profit from January to December with a line plot.

```python
monthly_profit=data.groupby("Months")["Profit"].sum()
monthly_profit
#sb = data["Months_abrv"].drop_duplicates()
#monthly_profit["Month_abrv"] = sb
months = data["Months_abrv"].drop_duplicates()#
```

```python
plt.rcdefaults()
plt.figure(figsize=[12,8])
plt.grid()
plt.title("Trend of profit from January to December ", fontsize = 16)
plt.xlabel("Months",fontsize=16)
plt.ylabel("Profit",fontsize=16)
plt.plot(months,monthly_profit)
```

## iii Which month does the company make more profit and what 5 products are sold most in that month?

### The company make more profit in November 

```python
#Top_products = data.groupby("Months_abrv"==M)["Product Name"]
Nov_products = pd.DataFrame(data.loc[data["Months_abrv"]=="Nov"])
Nov_Top_products = pd.DataFrame(Nov_products.groupby("Product Name")["Sales"].sum().sort_values().nlargest(5))
Nov_Top_products 
```

# Question 4

## i what is the correlation between sales and discount?

```python

#data["PRICE"].corr(data["RM"])
print(data["Discount"].describe())
corr_sales_discount = data["Discount"].corr(data["Sales"])#correlation between sales and discount
print(f"The correlation between Discount and Sales is {round(corr_sales_discount,2)}")
```

### Analyzing the data

### There is low correlation between Discount and Sales (- 0.08) and  A change in discount will probably not have significant effect on the Sales.

## II  Use a scatter plot to show the relationship between sales and profit.

```python

sns.set(rc = {'figure.figsize':(15,8)},font_scale = 1.6)
sns.set_style("white")
sns.scatterplot(data=data, x="Sales", y="Profit", hue="Category").set(title='scatter plot to show the relationship between sales and profit')
```

# Question 5

## i. A grouped bar chart showing the total profit for the different product category for each market.

```python
cat_Mak_profit = data.groupby(["Category","Market"])["Profit"].sum()
```

```python

sns.set(rc = {'figure.figsize':(15,8)},font_scale = 1.6)
sns.set_style("white")
ax = cat_Mak_profit.unstack(0).plot(kind="bar")#ploting the group bar chart
ax.set(title='Grouped bar chart showing the total profit for the different product category for each market.')
ax.set_ylabel("Profit", fontsize = 25)
```
